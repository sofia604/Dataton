## **Librerías**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/exportaciones.csv', sep=';')
```

```python
df
```

## Preprocesado de datos

```python
df
```

El dataset presentado contiene la siguiente información.
    - AÑO y MES: De cuando se realizó en envío.
    - SECTOR: Tipo de mercancía transportada.
    - PAIS: Destino de la mercancía.
    - FOB: Siglas de 'free on board'. Valor en USD de la mercancía incluyendo el coste de movilización y    cargos de aduana.

```python
# Se llevan las columnas de año y mes a formato de fecha
df['FECHA'] = df['MES'].astype("str") + "-" + df['AÑO'].astype("str")
df['FECHA'] = pd.to_datetime(df['FECHA'])
df.drop(columns=['AÑO', 'MES'], inplace=True)
```

```python
df
```

## Agrupación de datos

### Valor de las exportaciones por mes

```python
df_mensual = df[["AÑO","MES","FOB (MILES USD)"]].groupby(["AÑO","MES"],as_index=False).sum()
```

```python
df_mensual['FECHA'] = df_mensual['MES'].astype("str") + "-" + df_mensual['AÑO'].astype("str")
df_mensual['FECHA'] = pd.to_datetime(df_mensual['FECHA'])
df_mensual.drop(columns=['AÑO', 'MES'], inplace=True)
```

```python
df_mensual = df_mensual.set_index("FECHA")
```

```python
df_mensual.plot(figsize=(12,8))
plt.show()
```

## **Predicción**

```python
df_mensual
```

```python
df_mensual2 = df_mensual
```

```python
df_mensual2['FOB_LastMonth']=df_mensual2['FOB (MILES USD)'].shift(+1)
df_mensual2['FOB_2Monthsback']=df_mensual2['FOB (MILES USD)'].shift(+2)
df_mensual2['FOB_3Monthsback']=df_mensual2['FOB (MILES USD)'].shift(+3)
df_mensual2
```

```python
df_mensual2=df_mensual2.dropna()
df_mensual2
```

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_features=3, random_state=1)
```

```python
import numpy as np
x1,x2,x3,y=df_mensual2['FOB_LastMonth'],df_mensual2['FOB_2Monthsback'],df_mensual2['FOB_3Monthsback'],df_mensual2['FOB (MILES USD)']
x1,x2,x3,y=np.array(x1),np.array(x2),np.array(x3),np.array(y)
x1,x2,x3,y=x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
final_x=np.concatenate((x1,x2,x3),axis=1)
print(final_x)
```

```python
X_train,X_test,y_train,y_test=final_x[:-30],final_x[-30:],y[:-30],y[-30:]
```

```python
model.fit(X_train,y_train)
lin_model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)
plt.plot(pred,label='Random_Forest_Predictions')
plt.plot(y_test,label='Actual Sales')
plt.legend(loc="upper left")
plt.show()
```

```python
lin_pred=lin_model.predict(X_test)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11,6)
plt.plot(lin_pred,label='Linear_Regression_Predictions')
plt.plot(y_test,label='Actual Sales')
plt.legend(loc="upper left")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse_rf=sqrt(mean_squared_error(pred,y_test))
rmse_lr=sqrt(mean_squared_error(lin_pred,y_test))
```

```python
print('Mean Squared Error for Random Forest Model is:',rmse_rf)
print('Mean Squared Error for Linear Regression Model is:',rmse_lr)
```

```python
from pmdarima.arima import auto_arima
```

```python
#Conjunto de prueba y testeo
length_train = int(df_mensual.shape[0]*0.8)
train = df_mensual.iloc[:length_train,:]
test=df_mensual.iloc[length_train:,: ]
```

```python
arima_model = auto_arima(train, start_p=0, d=1, start_q=0,
                         max_p=5, max_d=4, max_q=5, start_P=0,
                         D=1, start_Q=0, max_P=5, max_D=4,
                         max_Q=5, m=12, seasonal=True, error_action="warn", trace=True,
                         supress_warning=True, stepwise=True,
                         random_state=20, n_fits=34)
```

```python
arima_model.summary()
```

```python
#Forecasting on the test data
sarima = pd.DataFrame(arima_model.predict(n_periods=35), index=test.index)
sarima.rename(columns= {0:"predicted_values"}, inplace=True)
```

```python
sarima
```

```python
plt.figure(figsize=(8,5))
plt.plot(train, label="Training")
plt.plot(test, label="Testing")
plt.plot(sarima, label="predicted")
plt.legend(loc="Left corner")
plt.show()
```
